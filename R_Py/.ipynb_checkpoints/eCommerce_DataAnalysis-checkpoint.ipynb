## 1.3 Data wrangling

### Data imporation

```python
ecom_data_oct = pd.read_csv('Data/2019-Oct.csv')
ecom_data_nov = pd.read_csv('Data/2019-Nov.csv')
ecom_data_dec = pd.read_csv('Data/2019-Dec.csv')
ecom_data_jan = pd.read_csv('Data/2020-Jan.csv')
```

### Data merging

```python
ecom_data_list =[ecom_data_oct, ecom_data_nov, ecom_data_dec, ecom_data_jan]
ecom_data_merging = pd.concat(ecom_data_list)
print(ecom_data_merging.shape) #Dimension dataframes
ecom_data_merging.head(5) #View data source
```

### Data subset on purchased products 

```python
ecom_merging_purchase0 = ecom_data_merging.loc[ecom_data_merging.event_type == 'purchase']
print(ecom_merging_purchase0.shape) #Dimension dataframe

##Remove the duplicates based on user_id, user_session, product_id and event_type
ecom_merging_purchase = ecom_merging_purchase0.drop_duplicates(['event_type', 'product_id', 'user_id','user_session'])
print(ecom_merging_purchase.shape) #Dimension dataframe
ecom_merging_purchase.head(5)
```

### Data subset on not purchased products

```python
ecom_merging_not_purchase0 = ecom_data_merging.loc[ecom_data_merging.event_type != 'purchase']
print(ecom_merging_not_purchase0.shape) #Dimension dataframe

##Remove the duplicates based on user_id, user_session, product_id and event_type
ecom_merging_not_purchase = ecom_merging_not_purchase0.drop_duplicates(['event_type', 'product_id', 'user_id','user_session'])
print(ecom_merging_not_purchase.shape) #Dimension dataframe
ecom_merging_not_purchase.head(10)

##Manage duplicates "purchase" and "not purchase" in the sub dataset

##Reassigning all events type in "not purchase"
ecom_merging_not_purchase['event_type'] = "not purchase"
ecom_merging_not_purchase.head(5)
```

### Data merging between purchased and not purchased products

```python
ecom_data_final = pd.concat([ecom_merging_purchase, ecom_merging_not_purchase])
print(ecom_data_final.shape)
ecom_data_final.head(20)
```

### Exporting the final data merging into csv file

```python
ecom_data_final.to_csv("Outputs/ecom_data_cleaning.csv")
```
